```python
data_folder = './data/'
data_submission_folder = './data/submission_files/'
```

```python
files = os.listdir(data_submission_folder)
```

```python
num_files = 0
preds = np.zeros(40000)
for file in files:
    num_files += 1
    data_file = pd.read_csv(f'{data_submission_folder}{file}')
    if num_files == 1:
        final_data = data_file.copy()
    else:
        final_data.loc[:,file] = data_file.pred
```

```python
final_data
```

```python
train_labels = pd.read_csv(f'{data_folder}train_labels.csv')
```

```python
final_data_label = final_data.merge(train_labels,how='left',on='id')
```

```python
train_final = final_data_label[final_data_label.id < 20000]
test_final = final_data_label[final_data_label.id >= 20000]
```

```python
X = train_final.drop(columns='label')
y = train_final.label
```

```python
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
params = {"objective": "binary",
          "boosting_type": "gbdt",
          "learning_rate": 0.03,
          "verbosity": -1,
}
```

```python
params = {"objective": "binary",
 "boosting_type": "gbdt",
 'n_estimators': 100, 
 'learning_rate': 0.04367959535512664, 
 'num_leaves': 1400, 
 'max_depth': 5, 
 'min_data_in_leaf': 400, 
 'min_gain_to_split': 12.924046793663218, 
 'bagging_fraction': 0.9, 
 'bagging_freq': 3, 
 'feature_fraction': 0.30000000000000004}
```

```python
dtrain = lgb.Dataset(X_train, y_train)
dvalid = lgb.Dataset(X_test, y_test, reference=dtrain)
```

```python
dtrain = lgb.Dataset(X, y)
```

```python
clf = lgb.train(params, 
                 dtrain)
```

```python
clf = lgb.train(params, 
                 dtrain, 
                 num_boost_round=300, 
                 valid_sets=dvalid, 
                 verbose_eval=100,
                 early_stopping_rounds=100)
```

```python
y_pred = clf.predict(test_final.drop(columns='label'))
```
